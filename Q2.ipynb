## Loading data

```python
english = {}
with open('./europarl_raw/test.en') as f:
    for line in f:
        for word in line.split():
            if word in english.keys():
                english[word] += 1
            else:
                english[word] = 1
```

```python
german = {}
with open('./europarl_raw/test.de') as f:
    for line in f:
        for word in line.split():
            if word in german.keys():
                german[word] += 1
            else:
                german[word] = 1
```

## Counting no. words

```python
print('No. English words: ', sum(english.values()))
print('No. German words: ', sum(german.values()))

print('No. English word types: ', len(english.keys()))
print('No. German word types: ', len(german.keys()))
```

## Vocab sizes

```python
english_unks = [key for key, value in english.items() if value==1]
german_unks = [key for key, value in german.items() if value==1]
print('No. English <UNK>s: ', len(english_unks))
print('No. German <UNK>s: ', len(german_unks))

print('English vocab size: ', len(english.keys()) - len(english_unks) + 1)
print('German vocab size: ', len(german.keys()) - len(german_unks) + 1)

print('Total vocab size: ', len(english.keys()) - len(english_unks) + len(german.keys()) - len(german_unks) + 1)
```

## Shared words

```python
shared_words = [word for word, count in english.items() if word in german.keys() and count > 1 and german[word] > 1]
print(len(shared_words))
print(shared_words)
```
